# Exploratory Data Analysis of Login Data

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_json('logins.json')
```

Sorting DataFrame.

```python
df_sorted = df.sort_values(by='login_time')
df_sorted.head()
```

```python
counts = df_sorted.resample('15min', on='login_time').count()
#counts = counts.head()
fig, ax = plt.subplots(figsize=(800,10))
ax.plot(counts)
ax.xaxis.set_major_locator(plt.MaxNLocator(120))
ax.set_title('Login Counts per 15 min Interval')
ax.set_xlabel('Time')
ax.set_ylabel('Counts')
plt.show()
```
